```python
import requests
from bs4 import BeautifulSoup
import pandas as pd






    
    

    
    

```

```python
n = [] #name
y = []#year
r = [] #rank
cf = []#certificate
dr = []#duration
g = []#genre
rate = []#rating
desc = []#description

links = [
    'https://www.imdb.com/search/title/?groups=top_250&sort=user_rating,desc&ref_=adv_prv',
    'https://www.imdb.com/search/title/?groups=top_250&sort=user_rating,desc&start=51&ref_=adv_nxt',
    'https://www.imdb.com/search/title/?groups=top_250&sort=user_rating,desc&start=101&ref_=adv_nxt',
    'https://www.imdb.com/search/title/?groups=top_250&sort=user_rating,desc&start=151&ref_=adv_nxt',
    'https://www.imdb.com/search/title/?groups=top_250&sort=user_rating,desc&start=201&ref_=adv_nxt'
]



for link in links:
    page = requests.get(link)
    soup = BeautifulSoup(page.content,'html.parser')

    a = soup.find_all('div',class_='lister-item mode-advanced')
    
    for i in range(len(a)):
        
        content = a[i].find('div',class_='lister-item-content')
    
    
        text = content.find('a')
        n.append(text.text)
    
        year = content.find('span',class_='lister-item-year text-muted unbold')
        year = year.text.replace('(',"").replace(")","")
        y.append(year)
        rank = content.find('span',class_='lister-item-index unbold text-primary')
        rank = rank.text.replace('.',"")
        r.append(rank)
    
        print(text.text,year,rank)
    
    
    
        d = content.find('p',class_="text-muted")
        certificate = d.find('span',class_='certificate')
    
        duration = d.find('span',class_='runtime')
        genre=d.find('span',class_='genre')
        genre=genre.text.split(",")
    
        if certificate:
            print(certificate.text)
            cf.append(certificate.text)
        else:
            cf.append("None")
    
        print(duration.text)
        dr.append(duration.text)
        print(genre[0])
        g.append(genre[0])
    
        rating = content.find('strong')
        print(rating.text)
        rate.append(rating.text)
    
        c =content.find_all('p',class_='text-muted')
        description = c[1].text
        print(description)
        desc.append(description)
        
        
    
```

```python
imdb_data = {"name":n,
             "year":y,
             "rank":r,
             "certificate":cf,
             "duration":dr,
             "genre":g,
             "rating":rate,
             "description":desc
    
}

data = pd.DataFrame.from_dict(imdb_data)
data = data.reset_index(drop=True)

print(data.info())
print(data['description'][0])

data.to_csv('imdb_data.csv')
```
